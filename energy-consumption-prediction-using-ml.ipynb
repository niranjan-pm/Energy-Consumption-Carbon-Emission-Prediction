```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Create a new column for the total electricity generation
df['Total Electricity (TWh)'] = df['Electricity from fossil fuels (TWh)'] + df['Electricity from nuclear (TWh)'] + df['Electricity from renewables (TWh)']

# Calculate the percentage contribution of each energy source
df['Fossil Fuels (%)'] = (df['Electricity from fossil fuels (TWh)'] / df['Total Electricity (TWh)']) * 100
df['Nuclear (%)'] = (df['Electricity from nuclear (TWh)'] / df['Total Electricity (TWh)']) * 100
df['Renewables (%)'] = (df['Electricity from renewables (TWh)'] / df['Total Electricity (TWh)']) * 100

# Plot the changes in the energy mix over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Fossil Fuels (%)', label='Fossil Fuels', data=df)
sns.lineplot(x='Year', y='Nuclear (%)', label='Nuclear', data=df)
sns.lineplot(x='Year', y='Renewables (%)', label='Renewables', data=df)
plt.title('Changes in Energy Mix Over the Years')
plt.xlabel('Year')
plt.ylabel('Percentage of Total Electricity')
plt.legend()
plt.show()

# Identify countries that have transitioned to a higher percentage of renewable energy
transitioned_countries = df[df['Renewables (%)'] > df['Renewables (%)'].shift(1)].sort_values('Year')

# Display the countries that have successfully transitioned
print('Countries that have successfully transitioned to a higher percentage of renewable energy:')
print(transitioned_countries[['Entity', 'Year', 'Renewables (%)']])
```

**Imputing using FILLNA**

```python
import pandas as pd

# Assuming df is your DataFrame

columns_to_impute = ['Access to electricity (% of population)', 'Renewable-electricity-generating-capacity-per-capita',
                     'Renewable energy share in the total final energy consumption (%)', 'Electricity from nuclear (TWh)']

# Create a copy of the DataFrame to avoid modifying the original data
df_imputed = df.copy()

# Iterate over each column to impute missing values within each country
for column in columns_to_impute:
    # Create a mask to identify missing values for the current column
    mask = df_imputed[column].isnull()

    # Group by 'Entity' and fill missing values within each group using the mean
    df_imputed[column] = df_imputed.groupby('Entity')[column].apply(lambda x: x.fillna(x.mean() ) if x.isnull().any() else x)

# Check the resulting DataFrame with imputed values
df_imputed.isnull().sum()
```

# Feature Selection

Feature selection is an important process in machine learning pipeline, aiming at eliminating redundent features and select only a subset of features. The removed features has a negative impact on the accuracy of machine learning model. So feature selection not only enhance machine learning model accuracy, but also reduces computation complexity.

From the broader perspective, feature selection is the process that is used to tackle the well known problem of curse of dimension. we achieve this by dimension reduction through feature selection, other methods of dimension reduction includes Feature extraction algorithms such as PCA or LDA, however we will not use feature extraction in this assessment.

Here we use to find the optimal number of features using feature selection algorithms such as Recursive Feature Elimination and pearsons correlation find the most optimal feature space.

As we are required to build two predictive models, one for Energy Consumption predictions and the other for Carbon Emission Predictions , so we select different sets of optimal feature sets for both in this section.

```python
#Preparing input and outputs

import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import ExtraTreesClassifier

#Selecting input
x_features = df_knn_imp.drop(columns= ['Primary energy consumption per capita (kWh/person)', 'Value_co2_emissions_kt_by_country','Latitude', 'Longitude','Entity'] , axis=1)

#Selecting target (output) for Energy consumption model
y_energy = df_knn_imp[['Primary energy consumption per capita (kWh/person)']]

#Selecting target (output) for Co2 Emission Model
y_co2= df_knn_imp['Value_co2_emissions_kt_by_country']

```

# Feature Selection using RFE with linear Regression model or Decsion Tree

Recursive Feature Elimination (RFE) is an embedded feature selection method that leverages Machine learning algorthim to rank features based on the chosen evaluation metric. It iterates multiple times thats why recursive until it finds the most optimal feature subset.
We use RFECV from SKlearn library for find the optimal feature subset.


```python
#Using Recursive Feature Elimination method with Linear regression model

from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#split for Energy Consumption Model
x_train,x_test ,y_train, y_test = train_test_split(x_features, y_energy ,test_size=0.3, random_state= 0)
```

```python
#split for CO2 Prediction Model
x_train,x_test ,y_train, y_test = train_test_split(x_features, y_co2 ,test_size=0.3, random_state= 0)
```

```python
#Using Recursive Feature Elimination method with Linear regression model

lm = LinearRegression()
dm = DecisionTreeRegressor()

rfe1 = RFE(dm, n_features_to_select=6)   # RFE with 20 features

# Fit on train and test data with 20 features
X_train_new = rfe1.fit_transform(x_train, y_train)
X_test_new = rfe1.transform(x_test)

selected_energy_features = rfe1.ranking_

# Print the boolean results
print("Number of Features Selected:", rfe1.n_features_)
print("Ranking of Features Selected:", rfe1.ranking_)

print("Selected Features:", rfe1.support_)
print("Selected Features Names:", rfe1.feature_names_in_)

dm.fit(X_train_new, y_train)
predictions_rfe = dm.predict(X_test_new)

RMSE = np.sqrt(mean_squared_error(y_test, predictions_rfe))
R2 = r2_score(y_test, predictions_rfe)
print('R2:',R2,'RMSE:',RMSE)     #Output R2: 0.88 RMSE: 0.33
```

```python
#RFECV using Random forest regressor with Cross validation


from sklearn.feature_selection import RFECV

# use any other model you want here.
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()

# build the RFE with CV option.
rfe = RFECV(model, min_features_to_select = 3, step = 1 , cv=3)

# fit the RFE to our data.
selection  = rfe.fit(x_train, y_train)

# print the selected features.
print(x_train.columns[selection.support_])
```

```python
# Final Features for Energy Prediction are here.

print('Selected Features for Carbon Emission Prediction Model:')

#print('Selected Features for Energy Consumption Prediction Model:')
features_energy = list(x_train.columns[selection.support_])
features_energy


```

```python
import matplotlib.pyplot as plt

min_features_to_select = 3
n_scores = len(rfe.cv_results_["mean_test_score"])
plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Mean test accuracy")
plt.errorbar(
    range(min_features_to_select, n_scores + min_features_to_select),
    rfe.cv_results_["mean_test_score"],
    yerr=rfe.cv_results_["std_test_score"],
)
plt.title("Recursive Feature Elimination \n with correlated features")
plt.show()
```


The plot reveals a plateau effect in prediction accuracy scores when selecting 8 to 10 features showing similar accuracy against the optimal number of features (8-10) for our dataset.

The Recursive Feature Elimination (RFE) determines the final feature subset based on this optimal number of features. As we increase the number of features beyond 10, the accuracy reduces showing the adverse effect of overfitting or introduction of correlated features.
```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Loading the dataset


# Defining the countries for CO2 emissions forecasting
countries = ['India', 'China', 'Brazil']  # Including Brazil instead of Turkey

# Filtering the data for the selected countries
filtered_data = df[df['Entity'].isin(countries)]

# Function to fit ARIMA model and forecast for CO2 Emissions
def manual_arima_forecast_co2(country_data, order=(1, 1, 1)):
    # Extract the CO2 Emissions column
    co2_emissions = country_data['Value_co2_emissions_kt_by_country'].dropna()

    # Fit ARIMA model
    model = ARIMA(co2_emissions, order=order)
    model_fit = model.fit()

    # Forecast for the next 5 years (2021-2025)
    forecast = model_fit.forecast(steps=5)

    return forecast, model_fit

# Forecasting CO2 Emissions for each country
manual_results_co2 = {}
for country in countries:
    country_data = filtered_data[filtered_data['Entity'] == country].sort_values('Year')
    forecast, model_fit = manual_arima_forecast_co2(country_data)
    manual_results_co2[country] = {
        'forecast': forecast,
        'model_summary': model_fit.summary()
    }

# Visualizing the CO2 Emissions forecasts along with the historical data
plt.figure(figsize=(15, 10))

# Function to plot CO2 Emissions historical data and forecast
def plot_forecast_co2(country, ax):
    country_data = filtered_data[filtered_data['Entity'] == country].sort_values('Year')
    years = country_data['Year']
    co2_emissions = country_data['Value_co2_emissions_kt_by_country']

    # Plot historical data
    ax.plot(years, co2_emissions, label='Historical Data')

    # Forecast data
    forecast_years = np.arange(2021, 2026)
    forecast = manual_results_co2[country]['forecast']
    ax.plot(forecast_years, forecast, label='Forecast', linestyle='--')

    ax.set_title(f'CO2 Emissions Forecast for {country}')
    ax.set_xlabel('Year')
    ax.set_ylabel('CO2 Emissions (kt)')
    ax.legend()

# Plot for each country
for i, country in enumerate(countries):
    ax = plt.subplot(len(countries), 1, i+1)
    plot_forecast_co2(country, ax)

plt.tight_layout()
plt.show()
```
